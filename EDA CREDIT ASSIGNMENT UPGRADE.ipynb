# Analysis of NAME_INCOME_TYPE column

```python
def plot_defaulter_non_defaulter(EDA_DF1, TARGET, gender_col):
    
    defaulter = EDA_DF1[EDA_DF1[TARGET] == 1]
    not_defaulter = EDA_DF1[EDA_DF1[TARGET] == 0]

    plt.subplot(1,2,1)
    plt.title("Not paid loan on time")
    sns.countplot(x = gender_col, data = defaulter)
    plt.xticks(rotation = 90)

    plt.subplot(1,2,2)
    plt.title("Paid loan on time")
    sns.countplot(x = gender_col, data = not_defaulter)
    plt.xticks(rotation = 90)

    plt.tight_layout(pad = 9)
    plt.show()
```

```python
plot_defaulter_non_defaulter(EDA_DF1, 'TARGET', 'CODE_GENDER')
```

```python
plot_defaulter_non_defaulter(EDA_DF1, 'TARGET', 'FLAG_OWN_CAR')
```
